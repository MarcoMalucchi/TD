## **Task 5**

Obiettivo: Verificare la risposta in frequenza del filtro a campione, verificando che essa sia in accordo con quanto atteso, facendo pratica con lo script.

* 10 Hz: Sinusoide in ingresso con ampiezza a 6 V ed offset nullo, si registra un'attenuazione di circa 0.43 ed uno sfasamento apprezzabile. 

* 100 Hz: Con la stessa forma d'onda in ingresso (ma frequenza 100 Hz), osservo uno sfasamento minimo ed un'attenuazione di circa 0.68.

* 1 kHz: Si è mandata in ingresso la stessa forma d'onda con ampiezza di 3 V. Si oserva un guadagno approssimativamente di 0.67, anche in questo caso, come atteso, lo sfasamento è minimo.

* 10 kHz: Stessa forma d'onda del caso precedente (ovviamente con frequenza modificata), si misura un guadagno approssimativo di circa 0.32 ed un forte sfasamento.

Se lo sfasamento risulta in accordo qualitativo con le simulazioni (almeno in termini di valore assoluto), lo stesso non può essere affermato per il guadagno, per il quale, invece, convertendo in dB i valori ottenuti qualitativamente, si osserva un missmatch con le frequenze nella simulazione. Sovrapponendo la simulazione al fit effettuato nel tesk 7, il piccolo missmatch è effettivamente evidente.

Si noti inoltre, l'indipendenza, attesa, dei risultati dall'ampiezza della forma d'onda.

### **Forma d'onda quadra**

In questo caso la forma d'onda in uscita cambia a seconda della frequenza della quadra in ingresso: a base frequenze rispetto alla prima frequenza di taglio l'aziende del derivatore è dominante ed infatti la forma d'onda viene derivata (slope positiva del grafico del guadagno), mentre nel reggione di guadagno massimo, tra le due frequenze di taglio, ho una forma d'onda molto simile a quella in ingresso.

Infine pre freuenze maggiori della frequenza di taglio dell'integratore, l'azione di quest'ultimo è dominante e quindi la forma d'onda in uscita risulta essere pari all'integrale di quella in ingresso.

L'attenuazione del segnale al variare della frequenza, invece, è minore rispetto al caso sinusoidale, grazie alle presenza delle armoniche di ordine superiore nella quadra, che si sommano all'armonica fondamentale

### **Forma d'onda triangolare**

Per questa forma d'onda si registra una variazione della stessa meno sensibile ala variazione in frequenza, essenzialmente cambia solo a basse ed alte frequenze. In tali regimi si ritrovano le stesse caratteristiche della quadra, cioè derivazione a basse frequenze (vedo una pinna di squalo in uscita a 10 Hz) ed integrazione ad alte (archi di parabola concatenati a 10 kHz).


## **Task 7**

Il fit della funzione è stato effettuato attraverso la funzione $\texttt{nlinfit}$ di matlab e le acquisizioni sono state effettuate cno il codice $\texttt{sweepfreq.ipynb}$ coprendo circa 3 decadi, da 0.1 kHz a 0.1 MHz.

Gli errori sono stati stimati fissando una frequenza nel centro dell'intervallo a guadagno massimo e prendendo per essa diverse misure di ampiezza, per poi calcolarne la deviazione standard del campione; si osserva subito che tale procedura restituirà un risultato meno accurato di quello che potrebbe restituire, ad esempio, fare diverse acquisizioni nello stesso intervallo di frequenza e calcolarne la deviazione standard frequenza per frequenza (è chiaro che sono necessarie diverse acquisizioni, nell'ordine di un centinaio). Questo è dovuto al fatto che la risposta in frequenza dell'oscilloscopio dell'AD2 cambia al cambiare della frequenza, in particolare peggiora all'aumentare della frequenza, quindi è chiaro che i punti ad alte frequenze avvranno un errore associato chiaramente maggiore degli altri. Essenzialmente, con la procedura eseguita, si sono sottostimati gli errori, almeno dei dati ad alte frequenze.

## **Task 8**

Obiettivo: Campionare un dente di sega a 10 kHz a frequenze di campionamento diverse avvicinandosi sempre più al limite di Nyquist, superandolo per poter osservare l'aliasing nell'FT del segnale.

Sono stati effettuati 4 campionamenti del segnale:

1. $f_c = 1$ MSa/S, si è ottenuto il seguente ASD:

<img src='FFT_Es02_1MSa.pdf' Width=500>

Questa è una buona acquisizione che rispetta pienamente il criterio di Nyquist: tutte le armoniche sono ben visibili e l'altezza dei picchi rispetta qualitativamente l'andamento 1/n. È presente, però, una modulazione periodica dell'altezza dei picchi delle armoniche più alte dovuta al leakage spettrale, infatti l'acquisizione è stata fatta a $2^14$ punti, enfatizzando, quindi, il leakage, nel task successivo questa modulazione scomparirà del tutto, effettuando un'acquisizione più oculata.

2. $f_c = 38$ kSa/S, si è ottenuto il seguente ASD:

<img src='FFT_Es02_38kSa.pdf' Width=500>

Già a questa frequenza di campionamento l'aliasing diventa apprezzabile:

* Il picco centrale a 10 kHz coincide con l'armonica fondamentale del segnale.

* Invece i picchi ai suoi lati coincidono con le armoniche della replica negativa dello spettro traslato della frequenza di campionamento, sommate alle armoniche originali del segnale, questa spiega la loro irregolarità ed il loro totale discostamento dall'andamento 1/n atteso.

3. $f_c = 10$ kSa/S, si è ottenuto il seguente ASD:

<img src='FFT_Es02_10kSa.pdf' Width=500>

L'ASD di questo plot è di facile interpretazione, infatti, campionando il segnale con la sua stessa frequenza, ottengo un  punto per periodo, ovviamente sempre circa allo stesso livello, quindi attraverso l'oscilloscopio vedo un segnale in DC, la cui FT presenta un solo picco, proprio in corrispondenza dell'armonica nulla (relativa proprio all'offset del segnale).

4. $f_c = 10.5$ kSa/S, si è ottenuto il seguente ASD:

<img src='FFT_Es02_10-5kSa.pdf' Width=500>

Per questo campionamento l'interpretazione è leggermente più complicata, si osserva che:

* L'andamento dell'altezza dei picchi 1/n e circa riprodotto, questo significa che il signale riprodotto dai campionamenti ha la stessa forma d'onda del segnale in ingresso (rampa di ddp)

* La spaziatura dei picchi è a 500 Hz anzi che a 10 kHz, come visibile nel primo campionamento (quello accurato), ciò è dovuto al fatto che, campionando ad una frequenza così vicina a quella del segnale originale, ho pochi punti ogni periodo, quindi il segnale ottenuto dai campionamenti risulterà avere una frequenza minore (sarà rallentato), in più, in questo caso risulterà anche temporalmente invertito (ottengo una ramp down di potenziale). Quindi, l'ASD riportato è del tutto virtuale, essendo relativo ad un segnale apparente.
